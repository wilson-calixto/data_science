```python
from bokeh.plotting import figure, show, output_file

colormap = {'hardtop':'yelow', 'wagon':'blue', 'sedan':'green', 'hatchback':'red', 'convertible':'black'}

#:               , , , , .
# 'alfa-romero':'yelow', 'audi':'blue', 'bmw':'green', 'chevrolet':'red', 'dodge':'black', 'honda':'purple','isuzu':'brown', 'jaguar':'Orange'mazda, mercedes-benz, mercury,mitsubishi, nissan, peugot, plymouth, porsche,renault, saab, subaru, toyota, volkswagen, volvo
#colormap = {'std': 'red', 'turbo': 'green'}

colors = [colormap[x] for x in dataframe['body-style']]

p = figure(title = "Iris Morphology")
p.xaxis.axis_label = 'Weight'
p.yaxis.axis_label = 'horsepower'
p.circle(dataframe["length"], dataframe["horsepower"], color=colors, fill_alpha=0.2, size=10)

show(p)
```

```python
# :         , , , six, , , .
colormap = {'twelve':'yelow','eight':'black','six':'Orange','five':'blue', 'four':'green', 'three':'purple', 'two':'red'}

colors = [colormap[x] for x in dataframe['num-of-cylinders']]

p = figure(title = "Iris Morphology")
p.xaxis.axis_label = 'horsepower'
p.yaxis.axis_label = 'price'
p.circle(dataframe["horsepower"], dataframe["price"], color=colors, fill_alpha=0.2, size=10)

show(p)


#
```
